```python
#The distribution is right skewed because of presence of outliers which can lead to overfitting in model.
#To overcome this right-skewed distribution we can use log normal transformation technique to bring a normal distribution.
#1️⃣ What is a right-skewed distribution?

#A right-skewed (positively skewed) distribution has a long tail on the right side.

#Most values are small, but there are a few very large values.

#Example: Bank balances, salaries, or transaction amounts often have this shape.

#Problem: Many statistical models (like linear regression) assume normality of variables. Right-skewed data can affect model performance.

#2️⃣ What is log transformation?

#A log transformation applies the natural logarithm (or base 10, base 2) to each value:

#xnew​=log(x)

#For example:

#Original values: [10, 50, 200, 1000]

#Log-transformed (base 10): [1, 1.7, 2.3, 3]

#Large values shrink more than small values → reduces skewness.

#3️⃣ Why use log transformation?

#Makes data more “normal”

#The histogram looks more symmetric.

#Reduces the effect of outliers

#Very large values no longer dominate the data.

#Improves model performance

#Models like linear regression or logistic regression perform better if predictors are approximately normal.
```

```python
conditions = [
    (df["NumOfProducts"]==1), 
    (df["NumOfProducts"]==2), 
    (df["NumOfProducts"]>2)
]

values = ["One product","Two Products","More Than 2 Products"]

df["Total_Products"] = np.select(conditions, values, default="Unknown")
df.drop(columns="NumOfProducts", inplace=True, errors="ignore")  # avoid error if already dropped
```

```python
X = df.drop(columns=["Churned"])
y = df["Churned"]
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
print("Shape of x_train is:",x_train.shape)
print("Shape of x_test is: ",x_test.shape)
print("Shape of y_train is:",y_train.shape)
print("Shape of y_test is: ",y_test.shape)
```

```python
smt = SMOTE(random_state=42)
```

```python
x_train_resampled,y_train_resampled = smt.fit_resample(x_train,y_train)
print(x_train_resampled.shape ,y_train_resampled.shape)
```

```python
y_train_resampled.value_counts()
```

```python
#Both the categories in Target variable are now having equal number of records.
#So we can train the Model pn balanced records for both churn categories and make a predictive model with low bias.
```

```python
#🔹 What is a Decision Tree?

#A Decision Tree is a supervised machine learning algorithm used for both classification and regression tasks.

#It splits the dataset into smaller and smaller subsets based on the most significant feature at each step.

#The final result is a tree-like model where:

#Nodes = questions or decisions based on features (e.g., “Is Balance > 50k?”).

#Branches = outcomes of the decisions (Yes/No or condition results).

#Leaves = final predictions (e.g., “Churn = Yes” or “Loan Approved”).
```

```python
#What is param_grid?

#It’s a dictionary where:

#keys = hyperparameters of the model

#values = list of possible options to try

#GridSearchCV will try all combinations of these parameters to find the best-performing model.
```

```python
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier()
```

```python
#dtree = DecisionTreeClassifier()
```

```python
param_grid = {"max_depth":[3,4,5,6,7,8,9,10],
              "min_samples_split":[2,3,4,5,6,7,8],
              "min_samples_leaf":[1,2,3,4,5,6,7,8],
              "criterion":["gini","entropy"],
              "splitter":["best","random"],
              "max_features":["auto",None],
              "random_state":[0,42]}
```

```python
grid_search = GridSearchCV(dtree, param_grid, cv=5, n_jobs=-1)

grid_search.fit(x_train_resampled,y_train_resampled)
```

```python
best_parameters = grid_search.best_params_

print("Best Parameters for DecisionTree Model is:\n\n")
best_parameters
```

```python
dtree = DecisionTreeClassifier(**best_parameters)

dtree.fit(x_train_resampled,y_train_resampled)
```

```python
y_train_pred = dtree.predict(x_train_resampled)
y_test_pred = dtree.predict(x_test)

print("Accuracy Score of Model on Training Data is =>",round(accuracy_score(y_train_resampled,y_train_pred)*100,2),"%")
print("Accuracy Score of Model on Testing Data  is =>",round(accuracy_score(y_test,y_test_pred)*100,2),"%")
```

```python
print("F1 Score of the Model is =>",f1_score(y_test,y_test_pred,average="micro"))
print("Recall Score of the Model is =>",recall_score(y_test,y_test_pred,average="micro"))
print("Precision Score of the Model is =>",precision_score(y_test,y_test_pred,average="micro"))
```

```python
cm = confusion_matrix(y_test,y_test_pred)

plt.figure(figsize=(15,6))
sns.heatmap(data=cm, linewidth=.5, annot=True, fmt="g", cmap="Set1")
plt.title("Model Evaluation using Confusion Matrix",fontsize=20,pad=20,fontweight="black")
plt.ylabel("Actual Labels")
plt.xlabel("Predicted Labels")
plt.show()
```

```python
y_pred_proba = dtree.predict_proba(x_test)[:][:,1]

df_actual_predicted = pd.concat([pd.DataFrame(np.array(y_test), columns=["y_actual"])])
df_actual_predicted.index = y_test.index


fpr, tpr, thresholds = roc_curve(df_actual_predicted["y_actual"], y_pred_proba)
auc = roc_auc_score(df_actual_predicted["y_actual"], y_pred_proba)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f"AUC = {auc:.2f}",color="green")
plt.plot([0, 1], [0, 1], linestyle="--", color="black")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve",pad=20,fontweight="black")
plt.legend()
plt.show()
```

```python
#1. What is ROC Curve?

#ROC = Receiver Operating Characteristic.

#It is a plot that shows the trade-off between sensitivity (recall) and specificity.

#It helps us check how good a classifier is at distinguishing between classes (Churn vs No Churn) at different thresholds.

#👉 A classifier does not just output Yes/No, it often gives a probability (e.g., 0.87 = 87% chance of churn).
#We can vary the threshold (default is 0.5) to see how performance changes.
```
